```python
import numpy as np 
import skfuzzy as fuzz
from skfuzzy import control as ctie

food = ctie.Antecedent(np.arange(0,11, 1), 'food')
service = ctie.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctie.Consequent(np.arange(10, 31, 1), 'tip')

food['poor'] = fuzz.trimf(food.universe,[0, 0, 5])
food['average'] = fuzz.trimf(food.universe,[0, 5, 5])
food['good'] = fuzz.trimf(food.universe,[0, 5, 10])

service['poor'] = fuzz.trimf(service.universe,[0, 0, 0])
service['average'] = fuzz.trimf(service.universe,[0, 5, 5])
service['good'] = fuzz.trimf(service.universe,[5, 5, 10])

tip['less'] = fuzz.trimf(tip.universe,[0, 5, 10] )
tip['normal'] = fuzz.trimf(tip.universe,[10, 15, 20])
tip['much'] = fuzz.trimf(tip.universe,[20, 25, 30])
food.view()
service.view()
tip.view()

rule1 = ctie.Rule(food['poor']&service['poor'],tip['less'])
rule2 = ctie.Rule(food['poor']&service['average'],tip['less'])
rule3 = ctie.Rule(food['poor']&service['good'],tip['less'])

rule4 = ctie.Rule(food['good']&service['poor'],tip['less'])
rule5 = ctie.Rule(food['good']&service['average'],tip['normal'])
rule6 = ctie.Rule(food['good']&service['good'],tip['much'])

rule7 = ctie.Rule(food['average']&service['poor'],tip['less'])
rule8 = ctie.Rule(food['average']&service['average'],tip['normal'])
rule9 = ctie.Rule(food['average']&service['good'],tip['much'])

tipping_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping = ctie.ControlSystemSimulation(tipping_ctie)

tipping.input['food']= 6.5
tipping.input['service']= 9.8
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping)
     
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctie

time = ctie.Antecedent(np.arange(10,181,1),'time')
rice = ctie.Antecedent(np.arange(100,1501,1),'rice')
power = ctie.Consequent(np.arange(0,101,1),'power')

time['very fast'] = fuzz.trimf(time.universe,[10, 10, 20])
time['fast'] = fuzz.trimf(time.universe,[10, 10, 20])
time['normal'] = fuzz.trimf(time.universe,[10, 20 ,60])
time['slow'] = fuzz.trimf(time.universe,[30, 50, 60])
time['very slow'] = fuzz.trapmf(time.universe,[40, 60, 150, 170])

rice['lightly'] = fuzz.trimf(rice.universe,[100,100,400])
rice['light'] = fuzz.trimf(rice.universe,[100,200,300])
rice['normal'] = fuzz.trimf(rice.universe,[200,800,1000])
rice['heavy'] = fuzz.trimf(rice.universe,[600,1000,1400])
rice['very heavy'] = fuzz.trimf(rice.universe,[800,1300,1500])

power['very weak']=fuzz.trimf(power.universe,[0,10,30])
power['weak']=fuzz.trimf(power.universe,[10,30,60])
power['normal']=fuzz.trimf(power.universe,[20,60,80])
power['strong']=fuzz.trimf(power.universe,[30,60,100])
power['very strong']=fuzz.trimf(power.universe,[80,80,100])

time.view()
rice.view()
power.view()

rule1 = ctie.Rule(time['very fast']&rice['light'],power['normal'])
rule2 = ctie.Rule(time['fast']&rice['light'],power['normal'])
rule3 = ctie.Rule(time['normal']&rice['light'],power['normal'])
rule4 = ctie.Rule(time['slow']&rice['light'],power['weak'])
rule5 = ctie.Rule(time['very slow']&rice['light'],power['very weak'])


rule6 = ctie.Rule(time['very fast']&rice['light'],power['strong'])
rule7 = ctie.Rule(time['fast']&rice['light'],power['strong'])
rule8 = ctie.Rule(time['normal']&rice['light'],power['normal'])
rule9 = ctie.Rule(time['slow']&rice['light'],power['normal'])
rule10 = ctie.Rule(time['very slow']&rice['light'],power['weak'])

rule11 = ctie.Rule(time['very fast']&rice['heavy'],power['very strong'])
rule12 = ctie.Rule(time['fast']&rice['heavy'],power['very strong'])
rule13 = ctie.Rule(time['normal']&rice['heavy'],power['strong'])
rule14 = ctie.Rule(time['slow']&rice['heavy'],power['normal'])
rule15 = ctie.Rule(time['very slow']&rice['heavy'],power['normal'])

rule16 = ctie.Rule(time['very fast']&rice['normal'],power['very strong'])
rule17 = ctie.Rule(time['fast']&rice['normal'],power['strong'])
rule18 = ctie.Rule(time['normal']&rice['normal'],power['strong'])
rule19 = ctie.Rule(time['slow']&rice['normal'],power['normal'])
rule20 = ctie.Rule(time['very slow']&rice['normal'],power['weak']) 
 
rule21 = ctie.Rule(time['very fast']&rice['very heavy'],power['very strong'])
rule22 = ctie.Rule(time['fast']&rice['very heavy'],power['very strong'])
rule23 = ctie.Rule(time['normal']&rice['very heavy'],power['strong'])
rule24 = ctie.Rule(time['slow']&rice['very heavy'],power['strong'])
rule25 = ctie.Rule(time['very slow']&rice['very heavy'],power['normal'])



power_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powering = ctie.ControlSystemSimulation(power_ctie)

powering.input['rice']= 700
powering.input['time']= 30
powering.compute()
print(powering.output['power'])
power.view(sim=powering)
```
